```python
import os
import json
import pandas as pd
import mysql.connector
```

```python
#aggregate_transaction

p1="D:/capstone/phonepe/pulse/data/aggregated/transaction/country/india/state/"
ag_trans_list=os.listdir(p1)

col={"states":[],"years":[],"quarter":[],"trans_type":[],"trans_count":[],"trans_amount":[]}

for state in ag_trans_list:
    current_state=p1+state+"/"
    ag_year_list=os.listdir(current_state)

    for year in ag_year_list:
        current_year=current_state+year+"/"
        ag_file_list=os.listdir(current_year)

        for file in ag_file_list:
            current_file=current_year+file
            data=open(current_file,"r")


            A=json.load(data)

            for i in A["data"]['transactionData']:
                name=i['name']
                count=i['paymentInstruments'][0]['count']
                amount=i['paymentInstruments'][0]['amount']
                col["trans_type"].append(name)
                col["trans_count"].append(count)
                col["trans_amount"].append(amount)
                col["states"].append(state)
                col["years"].append(year)
                col["quarter"].append(int(file.strip(".json")))
```

```python
A
```

```python
col
```

```python
aggregate_transaction=pd.DataFrame(col)
aggregate_transaction['states']=aggregate_transaction['states'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggregate_transaction['states']=aggregate_transaction['states'].str.replace('-',' ')
aggregate_transaction['states']=aggregate_transaction['states'].str.title()
aggregate_transaction['states']=aggregate_transaction['states'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')


```

```python
#aggregate_users

p2="D:/capstone/phonepe/pulse/data/aggregated/user/country/india/state/"
ag_user_list=os.listdir(p2)


col1={"states":[],"years":[],"quarter":[],"Brands":[],"trans_count":[],"Percentage":[]}

for state in ag_user_list:
    current_state=p2+state+"/"
    ag_year_list=os.listdir(current_state)

    for year in ag_year_list:
        current_year=current_state+year+"/"
        ag_file_list=os.listdir(current_year)

        for file in ag_file_list:
            current_file=current_year+file
            data=open(current_file,"r")


            B=json.load(data)

            try:
                for i in B["data"]["usersByDevice"]:
                    brand=i['brand']
                    count=i['count']
                    percentage=i['percentage']
                    col1["Brands"].append(brand)
                    col1["trans_count"].append(count)
                    col1["Percentage"].append(percentage)
                    col1["states"].append(state)
                    col1["years"].append(year)
                    col1["quarter"].append(int(file.strip(".json")))

            except:
                pass
```

```python
#SQL CONNECTION

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="phonepe"
)
cur= conn.cursor()
```

```python
#table creation for aggregate_transaction

create_query1=''' create table if not exists aggregate_transaction(states varchar(255), years int, quarter int, trans_count bigint,
                                                                   trans_amount bigint)'''
cur.execute(create_query1)
conn.commit()

insert_query1='insert into aggregate_transaction(states,years,quarter,trans_type,trans_count,trans_amount) values (%s,%s,%s,%s,%s,%s)'

data=aggregate_transaction.values.tolist()


cur.executemany(insert_query1,data)
conn.commit()
```
